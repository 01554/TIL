```python
#  完全一致率を出す関数
def accuracy(outputs:list[str], labels:list[str]):
    num_match = sum(1 if t == p else 0 for t,p in zip(outputs,labels))
    return num_match / len(outputs)
```

```python
# prompt: F1 score  を算出する関数、渡されるデータはテキスト、中身はラベル郡で不定、 skleranは使わない

def f1_score(y_true, y_pred):
  """
  F1スコアを計算する関数。

  Args:
    y_true: 正解ラベルのリスト。
    y_pred: 予測ラベルのリスト。

  Returns:
    F1スコア。
  """

  true_positives = 0
  false_positives = 0
  false_negatives = 0

  for true_label, predicted_label in zip(y_true, y_pred):
    print(true_label, predicted_label)
    if true_label == predicted_label:
      true_positives += 1

    # そうか、文字を判定したい場合は　 １，０に収束させる　f1_score このままじゃだめじゃん、なんか別の方法ださなきゃ
    elif predicted_label == 1 and true_label == 0:
      false_positives += 1
    elif predicted_label == 0 and true_label == 1:
      false_negatives += 1
  print(true_positives, false_positives, false_negatives)

  precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
  recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
  print(precision, recall)
  f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

  return f1_score
```

```python
# prompt: 文字ベースの f1_score : 正解テキストと予測テキストの文字の一致に基づくf1_score

def char_f1_score(y_true, y_pred):
  """
  文字ベースのF1スコアを計算する関数。

  Args:
    y_true: 正解テキストのリスト。
    y_pred: 予測テキストのリスト。

  Returns:
    F1スコア。
  """

  true_positives = 0
  false_positives = 0
  false_negatives = 0

  for true_text, predicted_text in zip(y_true, y_pred):
    # 文字列を比較して共通の文字数をカウント
    common_chars = set(true_text) & set(predicted_text)
    true_positives += len(common_chars)

    # 予測テキストにのみ含まれる文字数をカウント
    false_positives += len(set(predicted_text) - set(true_text))

    # 正解テキストにのみ含まれる文字数をカウント
    false_negatives += len(set(true_text) - set(predicted_text))

  precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
  recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0

  f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

  return f1_score
```

```python
char_f1_score(['あいうえお'],['あいうえお'])
```

```python
char_f1_score(['あいうヱお'],['あいうえお'])
```

```python
import numpy as np
from scipy.stats import pearsonr

# サンプルデータ (例)
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 6]

# ピアソン積率相関係数を求める
correlation, p_value = pearsonr(x, y)

print(f"ピアソン積率相関係数: {correlation}")
print(f"p値: {p_value}")
```

```python
from scipy.stats import spearmanr

# サンプルデータ (例)
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 6]

# スピアマンの順位相関係数を求める
correlation, p_value = spearmanr(x, y)

print(f"スピアマンの順位相関係数: {correlation}")
print(f"p値: {p_value}")
```
